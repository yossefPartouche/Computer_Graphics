# Finding the Norm of a Vector

## Key Definitions

**Vector**: A mathematical object with both magnitude and direction, represented as an ordered set of components, e.g., $\mathbf{v} = (v_1, v_2, v_3)$ in 3D space.

**Norm (or Magnitude)**: The length of a vector, denoted as $\|\mathbf{v}\|$ or $|\mathbf{v}|$, representing the distance from the origin to the point defined by the vector.

**Euclidean Norm** (also called the L2 norm): For a vector in n-dimensional space, it's calculated as:
$\|\mathbf{v}\| = \sqrt{v_1^2 + v_2^2 + ... + v_n^2}$

## Method for Finding a Vector's Norm

1. **Identify the vector components**: Write out all component values clearly.
2. **Square each component**: Calculate the square of each value.
3. **Sum the squared values**: Add all squared components together.
4. **Take the square root**: Calculate the square root of the sum to get the norm.

### Example Problem

**What is norm of the following vector: v = (2, 3, -âˆš3) ?**

#### Step 1: Identify the components
The vector $\mathbf{v}$ has three components: $v_1 = 2$, $v_2 = 3$, and $v_3 = -\sqrt{3}$

#### Step 2: Square each component
$v_1^2 = 2^2 = 4$
$v_2^2 = 3^2 = 9$
$v_3^2 = (-\sqrt{3})^2 = 3$ (note that squaring removes the negative sign)

#### Step 3: Sum the squared components
$v_1^2 + v_2^2 + v_3^2 = 4 + 9 + 3 = 16$

#### Step 4: Take the square root
$\|\mathbf{v}\| = \sqrt{16} = 4$

#### Therefore:
The norm of vector $\mathbf{v} = (2, 3, -\sqrt{3})$ is 4.

#### Verification:
We can verify this is correct because $\|\mathbf{v}\| = \sqrt{2^2 + 3^2 + (-\sqrt{3})^2} = \sqrt{4 + 9 + 3} = \sqrt{16} = 4$

#### Practice Question
What is the norm of the following vector in 5-dimensional space: $\mathbf{v} = (3, -2, 1, 4, -2)$?

### Solution

##### Step 1: Square each component
$3^2 = 9$
$(-2)^2 = 4$
$1^2 = 1$
$4^2 = 16$
$(-2)^2 = 4$

##### Step 2: Sum the squared components
$9 + 4 + 1 + 16 + 4 = 34$

##### Step 3: Take the square root
$\|\mathbf{v}\| = \sqrt{34} \approx 5.83$

Therefore, the norm of vector $\mathbf{v} = (3, -2, 1, 4, -2)$ is $\sqrt{34}$.

#### Affine Space and Points

### What is an Affine Space?

An affine space is a mathematical structure that extends the concept of a vector space by explicitly distinguishing between points and vectors. Unlike a vector space where the origin is fixed, an affine space doesn't have a distinguished origin point.

#### Definition of Affine Space:
An affine space is a set $A$ together with a vector space $V$ and an operation $+: A \times V \rightarrow A$ (called the action of $V$ on $A$) satisfying:

1. For any point $p \in A$ and the zero vector $\vec{0} \in V$: $p + \vec{0} = p$
2. For any point $p \in A$ and vectors $\vec{v}, \vec{w} \in V$: $p + (\vec{v} + \vec{w}) = (p + \vec{v}) + \vec{w}$**$ $
3. For any point $p \in A$ and vector $\vec{v} \in V$, there exists a unique point $q \in A$ such that $p + \vec{v} = q$
4. For any two points $p, q \in A$, there exists a unique vector $\vec{v} \in V$ such that $p + \vec{v} = q$

$**$ note that we have to distinguish between '$+$' defined over $A$ and '+' defined over $V$ similar concept to $0_v$ and $0_F$.

#### Definition of a Point in Affine Space:
A point in affine space is an element of the set $A$ in the definition above. Points represent positions in space without a specific reference to the origin.

### Differences Between Points, Vectors, and Scalars

#### Points:
- Represent positions or locations in affine space
- Cannot be added to other points (no point + point operation)
- Cannot be multiplied by scalars
- The difference of two points yields a vector: $q - p = \vec{v}$ (where $\vec{v}$ is the vector from $p$ to $q$)
- No natural "zero" point exists in affine space

#### Vectors:
- Represent displacements or directions in vector space
- Can be added to other vectors: $\vec{v} + \vec{w}$
- Can be multiplied by scalars: $\alpha \vec{v}$
- Have a natural zero element: $\vec{0}$
- Apply to points via the action: $p + \vec{v}$ (yielding another point)

#### Scalars:
- Elements of a field (usually real numbers)
- Used to scale vectors but not points
- Can be added and multiplied with each other
- Act on vectors through scalar multiplication

### Key Relationship:
The fundamental relationship in affine geometry is that while we can't meaningfully add two points, we can:
1. Add a vector to a point to get another point: $p + \vec{v} = q$
2. Subtract two points to get the vector between them: $q - p = \vec{v}$

This formalism is crucial in computer graphics where we need to precisely distinguish between locations (points) and displacements/directions (vectors).

#### Euclidean Space $E^n$

#### Formal Definition

**Euclidean space** $E^n$ is an $n$-dimensional real vector space equipped with an inner product and the associated Euclidean norm, making it a complete metric space.

#### Key Components:

1. **Set Structure**: $E^n = \mathbb{R}^n$, the set of all ordered $n$-tuples of real numbers $(x_1, x_2, ..., x_n)$

2. **Vector Space Properties**:
   - Addition: $(x_1, x_2, ..., x_n) + (y_1, y_2, ..., y_n) = (x_1 + y_1, x_2 + y_2, ..., x_n + y_n)$
   - Scalar multiplication: $\alpha(x_1, x_2, ..., x_n) = (\alpha x_1, \alpha x_2, ..., \alpha x_n)$
   - Zero vector: $(0, 0, ..., 0)$

3. **Inner Product**: $\langle \mathbf{x}, \mathbf{y} \rangle = \sum_{i=1}^n x_i y_i = x_1y_1 + x_2y_2 + ... + x_ny_n$

4. **Norm**: $\|\mathbf{x}\| = \sqrt{\langle \mathbf{x}, \mathbf{x} \rangle} = \sqrt{\sum_{i=1}^n x_i^2}$

5. **Distance**: $d(\mathbf{x}, \mathbf{y}) = \|\mathbf{x} - \mathbf{y}\| = \sqrt{\sum_{i=1}^n (x_i - y_i)^2}$

#### Essential Properties:

- **Completeness**: Every Cauchy sequence of points converges to a point in the space
- **Metric Structure**: The distance function satisfies the axioms of a metric:
  - Non-negativity: $d(\mathbf{x}, \mathbf{y}) \geq 0$
  - Identity of indiscernibles: $d(\mathbf{x}, \mathbf{y}) = 0$ if and only if $\mathbf{x} = \mathbf{y}$
  - Symmetry: $d(\mathbf{x}, \mathbf{y}) = d(\mathbf{y}, \mathbf{x})$
  - Triangle inequality: $d(\mathbf{x}, \mathbf{z}) \leq d(\mathbf{x}, \mathbf{y}) + d(\mathbf{y}, \mathbf{z})$

#### Connection to Your Sphere Modeling:

In the context of your notebook on implicit modeling, Euclidean space $E^3$ provides the mathematical foundation for representing spheres and other 3D objects. The sphere equation $(x-x_0)^2 + (y-y_0)^2 + (z-z_0)^2 = r^2$ defines all points in $E^3$ that are exactly distance $r$ from the center point $(x_0, y_0, z_0)$, using the Euclidean norm.

# Finding Points in an Affine Subspace

## Key Definitions:

- **Span**: The set of all linear combinations of vectors. For vectors $\mathbf{v}_1, \mathbf{v}_2, ..., \mathbf{v}_n$, the span is $\{\alpha_1\mathbf{v}_1 + \alpha_2\mathbf{v}_2 + ... + \alpha_n\mathbf{v}_n \mid \alpha_i \in \mathbb{R}\}$.

- **Affine Subspace**: A translation of a vector subspace. If $V$ is a vector subspace and $p$ is a point, then $A = p + V = \{p + v \mid v \in V\}$ is an affine subspace.

## Method for Solving:

### Step 1: Understand the Problem
- We have a vector subspace $V = \text{span}\{(1,2,0), (2,1,1)\}$
- We have a point $p = (1,2,3)$
- We want points $q$ in the affine subspace $A = p + V$

### Step 2: Find the General Form of Vectors in V
Any vector $v \in V$ can be written as:
$v = \alpha(1,2,0) + \beta(2,1,1)$ where $\alpha, \beta \in \mathbb{R}$
$v = (\alpha + 2\beta, 2\alpha + \beta, \beta)$

### Step 3: Find the General Form of Points in A
Points in $A$ have the form $q = p + v$ where $v \in V$:
$q = (1,2,3) + (\alpha + 2\beta, 2\alpha + \beta, \beta)$
$q = (1 + \alpha + 2\beta, 2 + 2\alpha + \beta, 3 + \beta)$

### Step 4: Check if Given Point is in A
For a given point $(x,y,z)$, we need to determine if there exist values of $\alpha$ and $\beta$ such that:
- $x = 1 + \alpha + 2\beta$
- $y = 2 + 2\alpha + \beta$
- $z = 3 + \beta$

This gives us a system of 3 equations with 2 unknowns. For the point to be in $A$, this system must have a solution.

### Step 5: Solve for $\alpha$ and $\beta$
From the third equation: $\beta = z - 3$
Substituting into the other equations:
- $x = 1 + \alpha + 2(z - 3) = 1 + \alpha + 2z - 6 = \alpha + 2z - 5$
- $y = 2 + 2\alpha + (z - 3) = 2 + 2\alpha + z - 3 = 2\alpha + z - 1$

Therefore:
- $\alpha = x - 2z + 5$
- $2\alpha = y - z + 1$

For consistency, we need: $2(x - 2z + 5) = y - z + 1$
Simplifying: $2x - 4z + 10 = y - z + 1$
Further: $2x - y - 3z + 9 = 0$

## Example Solution:

For the point $(3,5,4)$, check if it's in A:
- $2(3) - 5 - 3(4) + 9 = 6 - 5 - 12 + 9 = -2 \neq 0$
Therefore, $(3,5,4)$ is not in A.

For the point $(4,7,4)$, check:
- $2(4) - 7 - 3(4) + 9 = 8 - 7 - 12 + 9 = -2 \neq 0$
Therefore, $(4,7,4)$ is not in A.

For the point $(5,8,4)$, check:
- $2(5) - 8 - 3(4) + 9 = 10 - 8 - 12 + 9 = -1 \neq 0$
Therefore, $(5,8,4)$ is not in A.

For the point $(7,9,3)$, check:
- $2(7) - 9 - 3(3) + 9 = 14 - 9 - 9 + 9 = 5 \neq 0$
Therefore, $(7,9,3)$ is not in A.

For the point $(2,4,3)$, check:
- $2(2) - 4 - 3(3) + 9 = 4 - 4 - 9 + 9 = 0$ âœ“
Since the equation is satisfied, $(2,4,3)$ is in A.

Therefore, the point $(2,4,3)$ is in the affine subspace A.

### Cross Product: Definition and Computation

### Definition

The **cross product** (or vector product) is a binary operation on two vectors that produces a third vector perpendicular to both input vectors. It's denoted by the symbol Ã— between two vectors.

#### Properties of Cross Product
- **Anti-commutativity**: $\vec{a} \times \vec{b} = -(\vec{b} \times \vec{a})$
- **Distributivity**: $\vec{a} \times (\vec{b} + \vec{c}) = \vec{a} \times \vec{b} + \vec{a} \times \vec{c}$
- **Scalar multiplication**: $(\alpha\vec{a}) \times \vec{b} = \alpha(\vec{a} \times \vec{b})$
- **Perpendicularity**: $\vec{a} \times \vec{b}$ is perpendicular to both $\vec{a}$ and $\vec{b}$
- **Magnitude**: $|\vec{a} \times \vec{b}| = |\vec{a}||\vec{b}|\sin\theta$, where $\theta$ is the angle between vectors

#### Cross Product in $E^2$ (2D space)

In $E^2$, the traditional cross product is not defined in the same way as in $E^3$ because the result would be outside the 2D plane.

However, a 2D analogue exists that produces a scalar:

For vectors $\vec{a} = (a_x, a_y)$ and $\vec{b} = (b_x, b_y)$:
$$\vec{a} \times \vec{b} = a_x b_y - a_y b_x$$

This scalar represents the signed area of the parallelogram formed by the two vectors.

#### Cross Product in $E^3$ (3D space)

For vectors $\vec{a} = (a_1, a_2, a_3)$ and $\vec{b} = (b_1, b_2, b_3)$ in $E^3$:

$$\vec{a} \times \vec{b} = (a_2b_3 - a_3b_2, a_3b_1 - a_1b_3, a_1b_2 - a_2b_1)$$

#### Computation Methods:

##### 1. Component Formula
$$\vec{a} \times \vec{b} = 
\begin{pmatrix} 
a_2b_3 - a_3b_2 \\
a_3b_1 - a_1b_3 \\
a_1b_2 - a_2b_1
\end{pmatrix}$$

##### 2. Determinant Method
$$\vec{a} \times \vec{b} = 
\begin{vmatrix} 
\vec{i} & \vec{j} & \vec{k} \\
a_1 & a_2 & a_3 \\
b_1 & b_2 & b_3
\end{vmatrix}$$

Where:
- $\vec{i}$, $\vec{j}$, and $\vec{k}$ are unit vectors along the x, y, and z axes

#### Example:
For vectors $\vec{a} = (2, 3, 4)$ and $\vec{b} = (5, 6, 7)$:

$\vec{a} \times \vec{b} = 
\begin{pmatrix} 
3 \cdot 7 - 4 \cdot 6 \\
4 \cdot 5 - 2 \cdot 7 \\
2 \cdot 6 - 3 \cdot 5
\end{pmatrix} = 
\begin{pmatrix} 
21 - 24 \\
20 - 14 \\
12 - 15
\end{pmatrix} = 
\begin{pmatrix} 
-3 \\
6 \\
-3
\end{pmatrix}$

#### Geometric Interpretation
The cross product vector:
- Has magnitude equal to the area of the parallelogram formed by the two vectors
- Points in the direction determined by the right-hand rule
- Is zero if and only if the vectors are parallel (or one is zero)

This operation is crucial in computer graphics for calculating surface normals, which are essential for lighting calculations and visibility determination in the sphere modeling.

### Visual Interpretation of Cross Product

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.patches import Polygon

# Create figure with two subplots side by side
fig = plt.figure(figsize=(12, 5))
fig.suptitle('Cross Product Visualization', fontsize=14)

# ---- E^2 Cross Product (Left) ----
ax1 = fig.add_subplot(121)

# Define two 2D vectors
a_2d = np.array([3, 1])
b_2d = np.array([1, 2])

# Calculate cross product (scalar in 2D)
cross_product_2d = a_2d[0]*b_2d[1] - a_2d[1]*b_2d[0]

# Plot coordinate system
ax1.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax1.axvline(x=0, color='k', linestyle='-', alpha=0.3)

# Plot vectors
ax1.arrow(0, 0, a_2d[0], a_2d[1], head_width=0.2, head_length=0.3, fc='blue', ec='blue')
ax1.arrow(0, 0, b_2d[0], b_2d[1], head_width=0.2, head_length=0.3, fc='red', ec='red')

# Create parallelogram
vertices = np.array([[0, 0], a_2d, a_2d+b_2d, b_2d])
p = Polygon(vertices, alpha=0.2, color='green')
ax1.add_patch(p)

# Annotate vectors and area
ax1.text(a_2d[0]/2 - 0.3, a_2d[1]/2, '$\\vec{a}$', fontsize=12)
ax1.text(b_2d[0]/2, b_2d[1]/2 + 0.2, '$\\vec{b}$', fontsize=12)
ax1.text((a_2d[0]+b_2d[0])/2, (a_2d[1]+b_2d[1])/2, f'Area = {abs(cross_product_2d)}', fontsize=10)

ax1.set_xlim(-1, 5)
ax1.set_ylim(-1, 3.5)
ax1.set_aspect('equal')
ax1.grid(True, alpha=0.3)
ax1.set_title(f'Cross Product in $E^2$ (Scalar)\n$\\vec{{a}} \\times \\vec{{b}} = {cross_product_2d}$', fontsize=11)

# Add legend
ax1.plot([], [], 'b-', label='$\\vec{a}=(3,1)$')
ax1.plot([], [], 'r-', label='$\\vec{b}=(1,2)$')
ax1.legend(loc='upper right', fontsize=9)

# ---- E^3 Cross Product (Right) ----
ax2 = fig.add_subplot(122, projection='3d')

# Define two 3D vectors
a_3d = np.array([2, 0, 0])
b_3d = np.array([0, 2, 0])

# Calculate cross product (vector in 3D)
cross_product_3d = np.cross(a_3d, b_3d)

# Plot coordinate system
ax2.quiver(0, 0, 0, 3, 0, 0, color='gray', alpha=0.3, arrow_length_ratio=0.1)
ax2.quiver(0, 0, 0, 0, 3, 0, color='gray', alpha=0.3, arrow_length_ratio=0.1)
ax2.quiver(0, 0, 0, 0, 0, 3, color='gray', alpha=0.3, arrow_length_ratio=0.1)

# Plot vectors
ax2.quiver(0, 0, 0, a_3d[0], a_3d[1], a_3d[2], color='blue', arrow_length_ratio=0.1)
ax2.quiver(0, 0, 0, b_3d[0], b_3d[1], b_3d[2], color='red', arrow_length_ratio=0.1)
ax2.quiver(0, 0, 0, cross_product_3d[0], cross_product_3d[1], cross_product_3d[2], 
          color='green', arrow_length_ratio=0.1)

# Create parallelogram
ax2.plot_surface(
    np.array([[0, b_3d[0]], [a_3d[0], a_3d[0]+b_3d[0]]]),
    np.array([[0, b_3d[1]], [a_3d[1], a_3d[1]+b_3d[1]]]),
    np.array([[0, b_3d[2]], [a_3d[2], a_3d[2]+b_3d[2]]]),
    color='green', alpha=0.2
)

# Add text labels
ax2.text(a_3d[0]/2, a_3d[1]/2, a_3d[2]/2, '$\\vec{a}$', fontsize=10)
ax2.text(b_3d[0]/2, b_3d[1]/2, b_3d[2]/2, '$\\vec{b}$', fontsize=10)
ax2.text(cross_product_3d[0]/2, cross_product_3d[1]/2, cross_product_3d[2]/2 + 0.3, 
        '$\\vec{a} \\times \\vec{b}$', fontsize=10)

# Set labels and title
ax2.set_xlabel('X', fontsize=9)
ax2.set_ylabel('Y', fontsize=9)
ax2.set_zlabel('Z', fontsize=9)
ax2.set_title(f'Cross Product in $E^3$ (Vector)\n$\\vec{{a}} \\times \\vec{{b}} = (0, 0, 4)$', fontsize=11)
ax2.set_xlim(-0.5, 2.5)
ax2.set_ylim(-0.5, 2.5) 
ax2.set_zlim(-0.5, 4.5)

# Add text annotation explaining the result
""" 
ax2.text2D(0.05, 0.05, "Area of parallelogram = 4\nCross product points in z-direction\n(perpendicular to both vectors)",
          transform=ax2.transAxes, fontsize=9)
"""

# Make the figure's layout tight
plt.tight_layout()
plt.subplots_adjust(top=0.85)
plt.show()
```
