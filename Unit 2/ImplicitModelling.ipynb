A2. We Mathematically Represent the object (Sphere)
$$(x - x_0)^2 + (y - y_0)^2 + (z - z_0)^2 = r^2$$  
$$OR$$
$$(x - x_0)^2 + (y - y_0)^2 + (z - z_0)^2 - r^2 = 0$$  
Note: We're actually showing the boundary of the sphere in these equations

To use this formula we'll need:
1) Centre point of the sphere $(x_0, y_0, z_0)$
2) Radius $r$

### Step-by-Step Method for Finding the Radius of a Sphere from its Equation
When given an equation of a sphere in the form with mixed terms, follow these steps to find the radius:
#### Method:
1) Identify the standard form of a sphere equation: $(x - x_0)^2 + (y - y_0)^2 + (z - z_0)^2 = r^2$

2) Rearrange the given equation by completing squares for each variable:
- Group all terms with $x$
- Group all terms with $y$
- Group all terms with $z$
- Move constants to the right side
3) Complete the square for each variable group $^*$
4) Identify the center and radius from the rearranged equation

#### Identify the center and radius from the rearranged equation
#### Example:

For $x^2 + y^2 + z^2 -2x + 4y - 8z -4 = 0$

1) Rearrange terms  
$x^2 - 2x + y^2 + 4y + z^2 - 8z - 4= 0$

2) Group by variable  
$(x^2 - 2x) + (y^2 + 4y) + (z^2 - 8z) -4 = 0$

3) Complete the square for each variable  
For $x$ terms: $x^2 - 2x = (x - 1)^2 - 1$  
For $y$ terms: $y^2 + 4y = (y + 2)^2 - 4$  
For $z$ terms: $z^2 - 8z = (z - 4)^2 - 16$
4) Substitute back  
$(x - 1)^2 - 1 + (y + 2)^2 - 4 + (z - 4)^2 - 16 -4 = 0$

5) Rearrange to standard form  
$(x - 1)^2 + (y + 2)^2 + (z - 4)^2 = 1 + 4 + 16 + 4= 25$

6) Compare with standard form  
$(x - 1)^2 + (y + 2)^2 + (z - 4)^2 = r^2$

Therefore:

Center of sphere: $(1, -2, 4)$
Radius: $r = \sqrt{25} = 5$

#### Explicit vs. Implicit Formulas

##### Explicit Formulas
**Definition**: Directly expresses one variable in terms of others.

**Characteristics**:
- One variable isolated on left side
- Direct computation of output from input
- Function-like relationship

**Examples**:
1. $y = 3x + 2$ (line)
2. $z = \sqrt{x^2 + y^2}$ (cone)
3. $r = 5$ (sphere in spherical coordinates)

#### Implicit Formulas
**Definition**: Defines relationship between variables without isolating any variable.

**Characteristics**:
- All variables integrated in one equation
- Tests if points satisfy a condition
- Natural for representing surfaces

**Examples**:
1. $x^2 + y^2 + z^2 - 5^2 = 0$ (sphere)
2. $x^2 + y^2 - z = 0$ (paraboloid)
3. $x^2 - y^2 - z^2 = 1$ (hyperboloid)

### Converting Implicit to Explicit Form

When given an implicit formula for a 3D object, we often need to convert it to an explicit form z = f(x,y) for certain applications.

#### Step-by-Step Method:

1. **Isolate all terms with z variable** on one side of the equation

2. **Rearrange to standard form** with all z terms together

3. **Solve for z using algebra**
   - Complete the square for z terms if necessary
   - Apply quadratic formula if needed
   - Take square root when appropriate

4. **Determine the domain** where the explicit function is valid

#### Example:
Converting the implicit sphere equation: $x^2 + y^2 + z^2 - 2x + 4y - 8z - 4 = 0$

1) **Isolate z terms:**
   $z^2 - 8z = -(x^2 + y^2 - 2x + 4y - 4)$

2) **Complete the square for z:**
   $z^2 - 8z + 16 = -(x^2 + y^2 - 2x + 4y - 4) + 16$
   $(z - 4)^2 = 20 - x^2 - y^2 + 2x - 4y$

3) **Solve for z:**
   $z - 4 = \pm\sqrt{20 - x^2 - y^2 + 2x - 4y}$
   $z = 4 \pm\sqrt{20 - x^2 - y^2 + 2x - 4y}$

4) **Choose appropriate solution:**
   - For upper hemisphere: $z = 4 + \sqrt{20 - x^2 - y^2 + 2x - 4y}$
   - For lower hemisphere: $z = 4 - \sqrt{20 - x^2 - y^2 + 2x - 4y}$

5) **Simplify further (optional):**
   $z = 4 \pm\sqrt{25 - (x-1)^2 - (y+2)^2}$

The domain of this explicit function is the circle $(x-1)^2 + (y+2)^2 \leq 25$ on the xy-plane.

# Creating a Sphere Surface Using Explicit Equations

In computer graphics, we can use the explicit equation of a sphere to generate points on its surface and then connect them to create a mesh representation. Here's how we can do this:

## Using the Explicit Equation to Generate a Sphere

We have our explicit form from our earlier work:
$z = 4 \pm\sqrt{25 - (x-1)^2 - (y+2)^2}$

### Step 1: Sample the Domain:
What is "Sampling the Domain"?
Sampling the domain means selecting a finite set of points from the domain of our function to represent the continuous surface. For our sphere with explicit equation  
$z = 4 \pm\sqrt{25 - (x-1)^2 - (y+2)^2}$:
- Generate a grid of (x,y) points within the domain $(x-1)^2 + (y+2)^2 \leq 25$
- Typically use uniform sampling in a rectangular region, filtering points outside the domain

### Step 2: Calculate z-values
- For each valid (x,y) point, calculate:
  - Upper hemisphere: $z_{upper} = 4 + \sqrt{25 - (x-1)^2 - (y+2)^2}$
  - Lower hemisphere: $z_{lower} = 4 - \sqrt{25 - (x-1)^2 - (y+2)^2}$

### Step 3: Create Triangular Faces
- Connect adjacent points to form triangles
- Each set of three adjacent points forms a face
- Common patterns:
  1. Square grid: Each square becomes two triangles
  2. Radial pattern: Connect points in concentric circles

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.tri import Triangulation

# Step 0: Define sphere parameters
center_x, center_y, center_z = 1, -2, 4
radius = 5
grid_resolution = 15  # Lower for clarity in visualization

# Create figure with subplots to show each step - REDUCED SIZE
fig = plt.figure(figsize=(10, 8))  # Smaller figure size
fig.suptitle('Visualizing Sphere Creation Process', fontsize=14)

# Step 1: Sample the domain (show the grid and valid points)
# -----------------------------------------------------------
ax1 = fig.add_subplot(221)

# Create grid points
x = np.linspace(center_x - radius - 1, center_x + radius + 1, grid_resolution)
y = np.linspace(center_y - radius - 1, center_y + radius + 1, grid_resolution)
X, Y = np.meshgrid(x, y)

# Calculate which points are inside domain
inside_circle = (X - center_x)**2 + (Y - center_y)**2 <= radius**2

# Plot grid points
ax1.scatter(X, Y, c='lightgray', marker='.', s=15)  # Smaller markers
# Highlight points inside circle
ax1.scatter(X[inside_circle], Y[inside_circle], c='blue', marker='o', s=20)  # Smaller markers

# Draw the circle boundary
circle_theta = np.linspace(0, 2*np.pi, 100)
circle_x = center_x + radius * np.cos(circle_theta)
circle_y = center_y + radius * np.sin(circle_theta)
ax1.plot(circle_x, circle_y, 'r-')

# Mark the center point
ax1.scatter([center_x], [center_y], c='red', marker='x', s=60)  # Smaller marker

ax1.set_title('Step 1: Domain Sampling', fontsize=10)  # Smaller title
ax1.set_xlabel('X', fontsize=9)  # Smaller labels
ax1.set_ylabel('Y', fontsize=9)
ax1.set_aspect('equal')
ax1.grid(True, alpha=0.3)

# Step 2: Calculate Z values (show the points in 3D)
# -----------------------------------------------------------
ax2 = fig.add_subplot(222, projection='3d')

# Flatten the grid points that are inside the circle
x_valid = X[inside_circle]
y_valid = Y[inside_circle]

# Calculate z values for upper hemisphere only
z_valid = center_z + np.sqrt(radius**2 - (x_valid - center_x)**2 - (y_valid - center_y)**2)

# Plot the 3D points
ax2.scatter(x_valid, y_valid, z_valid, c='blue', marker='o', s=15)  # Smaller markers

# Add fewer vertical lines for clarity
for i in range(min(8, len(x_valid))):  # Reduced from 15 to 8 lines
    ax2.plot([x_valid[i], x_valid[i]], [y_valid[i], y_valid[i]], 
             [center_z, z_valid[i]], 'k--', alpha=0.3)

ax2.set_title('Step 2: Z-Value Calculation', fontsize=10)  # Smaller title
ax2.set_xlabel('X', fontsize=9)  # Smaller labels
ax2.set_ylabel('Y', fontsize=9)
ax2.set_zlabel('Z', fontsize=9)

# Step 3: Create triangular faces
# -----------------------------------------------------------
ax3 = fig.add_subplot(223, projection='3d')

# Use spherical coordinates for better triangulation
theta = np.linspace(0, np.pi/2, grid_resolution)  # Upper hemisphere only
phi = np.linspace(0, 2*np.pi, grid_resolution)

# Create 1D arrays of theta and phi for triangulation
theta_1d = np.repeat(theta, len(phi))
phi_1d = np.tile(phi, len(theta))

# Convert to Cartesian
x_sphere = center_x + radius * np.sin(theta_1d) * np.cos(phi_1d)
y_sphere = center_y + radius * np.sin(theta_1d) * np.sin(phi_1d)
z_sphere = center_z + radius * np.cos(theta_1d)

# Create triangulation in parameter space
tri = Triangulation(theta_1d, phi_1d)

# Plot wireframe to show triangulation
ax3.plot_trisurf(x_sphere, y_sphere, z_sphere, triangles=tri.triangles, 
                 alpha=0.8, color='blue', edgecolor='black', linewidth=0.3)  # Thinner lines

ax3.set_title('Step 3: Triangular Faces (Wireframe)', fontsize=10)  # Smaller title
ax3.set_xlabel('X', fontsize=9)  # Smaller labels
ax3.set_ylabel('Y', fontsize=9)
ax3.set_zlabel('Z', fontsize=9)

# Final result: The complete sphere
# -----------------------------------------------------------
ax4 = fig.add_subplot(224, projection='3d')

# Create a more detailed sphere for final visualization
resolution = 20  # Reduced from 30 for faster rendering
theta = np.linspace(0, np.pi/2, resolution)
phi = np.linspace(0, 2*np.pi, resolution)
theta, phi = np.meshgrid(theta, phi)

# Convert to Cartesian coordinates
x = center_x + radius * np.sin(theta) * np.cos(phi)
y = center_y + radius * np.sin(theta) * np.sin(phi)
z = center_z + radius * np.cos(theta)

# Plot the surface
surf = ax4.plot_surface(x, y, z, cmap='viridis', alpha=0.8, edgecolor='none')

ax4.set_title('Final Result: Smooth Hemisphere', fontsize=10)  # Smaller title
ax4.set_xlabel('X', fontsize=9)  # Smaller labels
ax4.set_ylabel('Y', fontsize=9)
ax4.set_zlabel('Z', fontsize=9)

plt.tight_layout()
plt.subplots_adjust(top=0.88, wspace=0.25, hspace=0.3)  # Adjusted spacing
plt.show()
```
