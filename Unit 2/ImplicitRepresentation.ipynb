## Implicit Representation  

We've seen in chapter 2.1 that the geometric representation of a physical ball was a sphere which had the implicit formula:  

$$(x - x_0)^2 + (y - y_0)^2 + (z - z_0)^2 - r^2 = 0$$ 

### General Implicit Forms

Implicit representations define shapes as the set of points where $f(\mathbf{x}) = 0$. The shape is the zero level set of the function.

#### Linear Functions

| Dimension | Form | Geometric Meaning |
|-----------|------|-------------------|
| 1D | $ax + b = 0$ | A point |
| 2D | $ax + by + c = 0$ | A line |
| 3D | $ax + by + cz + d = 0$ | A plane |

#### Quadratic Functions

| Dimension | Form | Geometric Meaning |
|-----------|------|-------------------|
| 1D | $ax^2 + bx + c = 0$ | Up to two points |
| 2D | $ax^2 + by^2 + cxy + dx + ey + f = 0$ | Conic sections (circle, ellipse, hyperbola, parabola) |
| 3D | $ax^2 + by^2 + cz^2 + dxy + exz + fyz + gx + hy + iz + j = 0$ | Quadric surfaces (sphere, ellipsoid, hyperboloid, paraboloid, cone) |

#### General Polynomial Functions

| Dimension | Form | Geometric Meaning |
|-----------|------|-------------------|
| 1D | $a_0 + a_1x + a_2x^2 + ... + a_nx^n = 0$ | Up to n points |
| 2D | $\sum a_{ij}x^iy^j = 0$ (where $i+j \leq n$) | Algebraic curve of degree n |
| 3D | $\sum a_{ijk}x^iy^jz^k = 0$ (where $i+j+k \leq n$) | Algebraic surface of degree n |

### Visualizing Level Sets

- **Inside**: Points where $f(\mathbf{x}) < 0$
- **Surface**: Points where $f(\mathbf{x}) = 0$
- **Outside**: Points where $f(\mathbf{x}) > 0$

For example, in a sphere centered at origin with radius $r$:
- Inside: $x^2 + y^2 + z^2 - r^2 < 0$
- Surface: $x^2 + y^2 + z^2 - r^2 = 0$
- Outside: $x^2 + y^2 + z^2 - r^2 > 0$

### Using LOF (Lower Order Explicit Function) to help represent Higher Order Geometry

- A 1D Explicit function can represent a curve in 2D such the graph of straight line and the graph of a parabola
- A 2D Explicit function can represent a surfave in 3D, where the domain (x,y) represents a plane and the range z represents the height at point(x,y)

### Limitations for using Explicit functions

1. Cannot have one to many mapping
2. Cannot bound a volume
3. Many surfaces cannot be defined explicitly using a simple formula

We'll now present these limitations over simple well known functions to gain intuition as to how we can overcome these obstacle using implicit formulas

Consider the equation of a line: $y ~=~ mx + c$

As we know that m represents the gradient of the line, which is equivelantly defiend by $m = tan(\theta)$ amount of change from the x-axis. Note that a vertical line means that $m ->∞$ to represent any line we must therfore use the implicit for $mx + by + c = 0$. If we did so we could represent a vertical line when $m ≠ 0, c = 0$.  

We'll change our perspective from a 1D implicit function to a 2D explicit function $f(x,y) = z = mx + by + c$  
So when we say $mx + by + c = 0$ we are saying $f(x,y) = z = 0$ which represents all the points on the 3D plane where the z coordinate is 0.

Consider the graph below: the XY (grey) plane is being intersected by the (purple) x + 2y + 3z -6 = 0 plane, notice that the intersection produces a line, which MUST LIE ON THE PLANE XY, this shown as the red line.

We conclude that changing the coefficients of m,b,c will change the line, and that any line can be represented using the implicit equation. In essence we constrained the 2D function to produce any 1D line.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a smaller figure
fig = plt.figure(figsize=(6, 5))  # Reduced from (10, 8)
ax = fig.add_subplot(111, projection='3d')

# Define the constants for the plane: ax + by + cz + d = 0
a, b, c, d = 1, 2, 3, -6

# Create a meshgrid for x and y values with fewer points
x_range = np.linspace(-5, 5, 15)  # Reduced from 20
y_range = np.linspace(-5, 5, 15)  # Reduced from 20
X, Y = np.meshgrid(x_range, y_range)

# Calculate z values from the plane equation: z = -(ax + by + d)/c
Z = -(a*X + b*Y + d)/c

# Find the intersection line with the XY plane (where z = 0)
x_line = np.linspace(-5, 5, 50)  # Reduced from 100
y_line = -(a*x_line + d)/b
z_line = np.zeros_like(x_line)  # z = 0 for the XY plane

# Plot the plane with simpler surface
plane = ax.plot_surface(X, Y, Z, alpha=0.5, color='blue', 
                        rstride=1, cstride=1, linewidth=0)

# Plot the XY plane (z=0)
ax.plot_surface(X, Y, np.zeros_like(X), alpha=0.2, color='gray')

# Plot the intersection line
ax.plot(x_line, y_line, z_line, 'r-', linewidth=2)  # Reduced from 3

# Set labels with smaller font
fontsize = 8  # Smaller font size
ax.set_xlabel('X', fontsize=fontsize)
ax.set_ylabel('Y', fontsize=fontsize)
ax.set_zlabel('Z', fontsize=fontsize)
ax.set_title(f'Plane {a}x + {b}y + {c}z + {d} = 0 intersecting XY plane', 
            fontsize=fontsize+2)

# Set axis limits
ax.set_xlim([-5, 5])
ax.set_ylim([-5, 5])
ax.set_zlim([-5, 5])

# Add a text label with smaller font
eq_text = f'Intersection: {a}x + {b}y + {d} = 0'
ax.text(0, -3, 0.5, eq_text, color='red', fontsize=fontsize)

# Adjust view angle
ax.view_init(elev=20, azim=30)

# Remove excess whitespace
plt.tight_layout(pad=0.5)

# Optionally, you can remove tick labels to save more space
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])

plt.show()
```

# Iso-Contour

An iso-contour (also called a level set or contour line) is a powerful concept in implicit representation that connects points of equal value in a scalar field.

### Definition

An iso-contour (or level set) for a scalar function $f$ is the set of all points where the function takes on a specified constant value $c$:

$$\{(x,y,z) \in \mathbb{R}^n \mid f(x,y,z) = c\}$$

### Iso-Contours in 2D

In two dimensions, for a scalar field $f(x,y)$:

- An iso-contour is a **curve** along which the function has a constant value $c$
- The formal definition is: $\{(x,y) \in \mathbb{R}^2 \mid f(x,y) = c\}$
- Common examples include:
  - Contour lines on topographic maps (connecting points of equal elevation)
  - Isobars on weather maps (connecting points of equal atmospheric pressure)
  - Isotherms (connecting points of equal temperature)

For example, in a 2D temperature distribution, each iso-contour connects all points that share the same temperature value.

## Iso-Contours in 3D

In three dimensions, for a scalar field $f(x,y,z)$:

- An iso-contour is a **surface** where the function has a constant value $c$
- The formal definition is: $\{(x,y,z) \in \mathbb{R}^3 \mid f(x,y,z) = c\}$
- Common examples include:
  - Equipotential surfaces in electromagnetic fields
  - Electron density surfaces in chemistry
  - Isosurfaces in medical imaging (CT/MRI scans)

For example, the surface of a sphere is an iso-contour of the distance function from the center point where $c = r$ (the radius).

## Importance in Computer Graphics

Iso-contours are fundamental in graphics because:

1. They provide a way to visualize scalar fields
2. They can extract meaningful boundaries from volumetric data
3. They are the basis of implicit surface modeling
4. The zero level set ($c = 0$) is typically used to define implicit surfaces
